### Преобразование значений уникальных номеров пользователей.

С помощью функции **_enumerate()_** создадим словарь соответствия индексов элементов, содержащих уникальное значение номера пользователя, уникальным номерам пользователей. Установим начальное значение индекса равное 1. Результат поместим в переменную **_dic_customer_id_**.

```python
dic_customer_id = dict(enumerate(df_store['customer_id'].unique(), 1))
```

Проверим наш словарь

```python
dic_customer_id
```

Создадим функцию **_change_id_**, которая принимает наш набор данных в качестве параметров. Внутри функции, для итерации номера  каждого элемента коллекции используем встроенную функцию **_map()_**.

```python
def change_id(df_store):
    df_store['customer_id'] = df_store['customer_id'].map(dic_customer_id)
```

В переменную **_'change_dic_customer_id'_** поместим словарь, который будет содержать два отдельных списка: список значений уникальных номеров пользователей и список индексов первого справочника, объединенных функцией **_zip()_**. Следующей строкой произведем замену длинных номеров на соответствующее значение индекса, с помощью функции **_map()_**.

```python
change_dic_customer_id = dict(zip(dic_customer_id.values(), dic_customer_id.keys()))
df_store['customer_id'] = df_store['customer_id'].map(change_dic_customer_id)
```

#### Распределение количества заказов по времени

Для проведения расчета метрик пользовательской активности, сгруппируем в отдельные таблицы набор данных по значению часа оформления заказа, даты оформления заказа, недели и месяца оформления заказа. Агрегацию данных заказов выполним методом **_nunique()_**, результат поместим:
* для сгруппированных по значению часа в переменную **_order_hau_**;
* для сгруппированных по значению даты в переменную **_order_dau_**
* для сгруппированных по значению недели в переменную **_order_wau_**
* для сгруппированных по значению месяца в переменную **_order_mau_**

Поскольку распределение заказов  в разрезе часа дает нам картину по всему интервалу времени (396 дней), для корректного отображения значений количества заказов в каждый час одного дня, разделим значения сгруппированной таблицы **_order_hau_** на количество дней временного интервала - 396.

Построим графики распределения метрик пользовательской активности:
1. **_'HAU'_** в разрезе часа;
2. **_'DAU'_** в разрезе дня;
3. **_'WAU'_** в разрезе недели;
4. **_'MAU'_** в разрезе месяца;

Для метрик пользовательской активности в отношении заказов

```python
order_hau = df_store.groupby('time').agg({'order_id': 'nunique'}).reset_index()
order_hau['order_id'] = order_hau['order_id'] / 396
order_dau = df_store.groupby('date').agg({'order_id': 'nunique'}).reset_index()
order_wau = df_store.groupby('week').agg({'order_id': 'nunique'}).reset_index()
order_mau = df_store.groupby('month').agg({'order_id': 'nunique'}).reset_index()
```

Повторим процедуру для метрик пользовательской активности в отношении номеров пользователей.

```python
cust_hau = df_store.groupby('time').agg({'customer_id': 'nunique'}).reset_index()
cust_hau['customer_id'] = cust_hau['customer_id'] / 396
cust_dau = df_store.groupby('date').agg({'customer_id': 'nunique'}).reset_index()
cust_wau = df_store.groupby('week').agg({'customer_id': 'nunique'}).reset_index()
cust_mau = df_store.groupby('month').agg({'customer_id': 'nunique'}).reset_index()
```

Для построения четырех парных графиков на одном полотне, создадим функцию **_metrics_plotting_**

```python
def metrics_plotting(df, column1, column2, title):
    df.plot(
        kind='line',
        x=column1, 
        y=column2, 
        figsize=(18,3), 
        color='gray',#'steelblue', 
        alpha=0.99, 
        linewidth=2, 
        legend=False,
        grid=True
    )
    plt.axhline(
        np.median(df[column2]),
        color='red', 
        linestyle=':', 
        linewidth=1.5, 
        label='median'
    )
    plt.title(
        'Метрика пользовательской активности: {} '.format(title), fontsize=14
    )
    leg = plt.legend(loc='best', fontsize=14)
    for l in leg.legendHandles:
        l.set_linewidth(4)
    plt.grid(True, linestyle=':')
    plt.show()
```

Строим графики метрик располагая для сравнения их попарно:

```python
metrics_plotting(order_hau, 'time', 'order_id', 'orders HAU')
metrics_plotting(cust_hau, 'time', 'customer_id', 'castomer HAU')

metrics_plotting(order_dau, 'date', 'order_id', 'orders DAU')
metrics_plotting(cust_dau, 'date', 'customer_id', 'castomer DAU')

metrics_plotting(order_wau, 'week', 'order_id', 'orders WAU')
metrics_plotting(cust_wau, 'week', 'customer_id', 'castomer WAU')

metrics_plotting(order_mau, 'month','order_id', 'orders MAU')
metrics_plotting(cust_mau, 'month', 'customer_id', 'castomer MAU')

hau = round((df_store.groupby('time').agg({'order_id': 'nunique'}).median() / 396), 2)
dau = df_store.groupby('date').agg({'order_id': 'nunique'}).median()
wau = df_store.groupby('week').agg({'order_id': 'nunique'}).median()
mau = df_store.groupby('month').agg({'order_id': 'nunique'}).median()

c_hau = round((df_store.groupby('time').agg({'customer_id': 'nunique'}).median() / 396), 2)
c_dau = df_store.groupby('date').agg({'customer_id': 'nunique'}).median()
c_wau = df_store.groupby('week').agg({'customer_id': 'nunique'}).median()
c_mau = df_store.groupby('month').agg({'customer_id': 'nunique'}).median()

print('orders HAU - cреднее количество заказов в час:  {}'.format(float(hau)))
print('castomer HAU - cреднее количество пользователей в час:  {}'.format(float(c_hau)))
print()
print('orders DAU - cреднее количество заказов в день:  {}'.format(int(dau)))
print('castomer DAU - cреднее количество пользователей в день:  {}'.format(int(c_dau)))
print()
print('orders WAU - cреднее количество заказов в неделю:  {}'.format(int(wau)))
print('castomer WAU - cреднее количество пользователей в неделю:  {}'.format(int(c_wau)))
print()
print('orders MAU - cреднее количество заказов в месяц:  {}'.format(int(mau)))
print('castomer MAU - cреднее количество пользователей в месяц:  {}'.format(int(c_mau)))
```

Сгруппируем набор данных **_df_store_** по значению месяца, агрегацию номеров заказов выполним методом **_count()_**, агрегацию номеров пользователей выполним методом **_nunique()_**, агрегацию товара в заказе  выполним методом **_nunique()_**, агрегацию значений количества позиций товара в заказе выполним методом **_count()_**, агрегацию цены товара  выполним методом **_sum()_**, агрегацию значений общей стоимости заказа выполним методом **_sum()_**, результат поместим в переменную **_monthly_profit_**. Добавим в набор **_monthly_profit_** столбец **_'avg_check'_**, в который поместим результат отношения общей стоимости заказов к количеству заказов и столбец **_'ltv'_**, в который поместим результат отношения общей стоимости заказов к количеству покупателей. 

```python
monthly_profit = df_store.groupby('month').agg(
    {'order_id': 'count', 
     'customer_id': 'nunique', 
     'product': 'nunique', 
     'quantity':'count',
     'price':'sum',
     'total_value':'sum'
     }
).reset_index().rename(
    columns={
        'order_id': 'orders',
        'customer_id': 'buyers',
        'product': 'products',
        'quantity':'quantitys',
        'price':'price_sum',
        'total_value':'revenue'
        }
    )
monthly_profit['avg_bill'] = (monthly_profit['revenue'] / monthly_profit['orders']).astype('int64')
monthly_profit['mean_profit'] = (monthly_profit['revenue'] / monthly_profit['buyers']).astype('int64')
```

Создадим функцию **_category_** для присвоения категории каждому товару.

```python
df['category'] = 0
def category(
    key_word_list, #словарь ключевых слов 
    category_name  #название категории
    ):
    join = '|'.join(key_word_list)
    index = df_store[df_store['key_word'].str.contains(join)].index.tolist()
    for i in index:
        df_store.loc[i, 'category'] = category_name
```

Категоризируем товары в нашем наборе, добавив в функции к имени словаря наименование категории

```python
category(kitchen, 'Товары для кухни')
category(textile, 'Текстиль и декор')
category(houseplants, 'Комнатные растения')
category(garden, 'Сад и огород')
category(bathroom, 'Товары для ванной')
category(household, 'Хозтовары')
category(storage, 'Уход и хранение')
category(bags, 'Сумки и тележки')
category(tools_accessories, 'Инструмент и метизы')
```

## Формирование сегментов покупателей

### Кластеризация покупателей

```python
df_store.head(3)
```

Нам поставлена задача сегментировать покупателей по профилю потребления.
Исходя из полученных в результате исследования данных, сервис носит разовый характер. Большинство пользователей сервисом совершили один заказ, среднее количество товаров в заказе не превышает 2 единиц, , средняя стоимость товара не превышает 500 ,значение среднего чека не превышает 700, общая стоимость заказа не превышает 1300. Учитывая слабые (практически отсутствующие) различия в числовых значениях данных (вышеперечисленных метрик), считаю, что использовать их в качестве признаков сегментирования менее предпочтительнее чем использовать категории товаров. 

Сегментируем покупателей на основе категории заказанных товаров.
При сегментации мы будем использовать только бинарные признаки наличия\отсутствия в заказе товаров той или иной категории.


Создадим копию нашего очищенного набора данных и поместим его в переменную **_cluster_data_**

```python
cluster_data = df_store.copy(deep=True)
```

Создадим сводную таблицу, в которой в качестве индекса используем номера пользователей, в качестве столбцов используем категории товаров, в качестве значений используем данные столбца **_'test_column'_**, агрегацию данных выполним методом **_mean()_**. Обновим индексацию и произведем замену пропусков на нули, результат поместим в переменную **_segmentation_**

```python
cluster_data.loc[:, 'test_column'] = 1

segmentation = cluster_data.pivot_table(
    index='customer_id', 
    columns = 'category', 
    values='test_column', 
    aggfunc='mean'
).reset_index().fillna(0)
```

```python
segmentation.head()
```

Зададим матрицу (объект-признак X и вектор целевой переменной y), применив метод **_drop()_** библиотеки **_Pandas_**: 

```python
X = segmentation.drop('customer_id', axis=1)
y = segmentation['customer_id']
```

Стандартизируем данные, создав объект класса **_scaler_**, затем обучим методом **_fit_** и преобразуем ~~обучающий и валидационные~~  методом **_transform_** набор

```python
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

Построим матрицу расстояний функцией **_linkage()_** на стандартизованной матрице признаков

```python
linked = linkage(X_sc, method = 'ward')
```

Строим дендрограмму

**Примечание**:

График строится 8 сек.

```python
#%%time
plt.figure(figsize=(15, 5))  
dendrogram(linked, orientation='top',
           truncate_mode='lastp',
           p=25,
           leaf_rotation=0,
           leaf_font_size=12,
           show_contracted = True ,)     
plt.title('Кластеризация покупателей', size=18)
plt.show()
```

Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры покупателей. Исходя из дендограммы примем число кластеров равное 9 (равное количеству категорий товара).

```python
km = KMeans(n_clusters = 9, random_state=10, n_jobs=1)# задаём число кластеров, равное 9
labels = km.fit_predict(X_sc)# применяем алгоритм к данным и формируем вектор кластеров
```

Cохраняем метки кластера в поле нашего датасета, поместив их в столбец **_'cluster_km'_**

```python
segmentation['cluster_km'] = labels
```

Посмотрим на средние значения признаков для кластеров

```python
clusters = segmentation.drop('customer_id', axis=1).groupby('cluster_km').mean().T
clusters
```

Построим тепловую матрицу

```python
plt.figure(figsize=(8, 5))
plt.title('Соотношение доли покупателей в кластерах с категориями товара', y=1.018, fontsize=16)
ax = sns.heatmap(clusters, alpha=0.9, annot = True, fmt='.2f', cbar=False,
            vmin=-0.80, vmax=0.79, cmap= 'coolwarm',
            linewidths=2, linecolor='gray',
            annot_kws={'weight': 'semibold', 'size': 12}
)
plt.tick_params(axis='both', labelsize=12)
plt.xlabel('Кластеры покупателей', fontsize=14)
plt.ylabel('Категории товара', fontsize=14)
plt.show()
```

Наблюдаем, что каждому из 9 кластеров характерна определенная категория товаров, при этом тепловая карта указывает на вхождении в корзину покупателей товаров разных категорий, например:
* покупатели, относящиеся к **6** кластеру (товары категории **_Сад и огород_**) приобретают 23 % товаров из категории **_Комнатные растения_** и около 7 %  **_Хозтоваров_**;
* покупатели, относящиеся к **8** кластеру (товары категории **_Инструмент и метизы_**) приобретают от 6 до 9 % из остальных восьми категорий товара;

Исходя из карты, в большинстве случаев покупатели делают покупки товаров одной категории, учтем это при подготовке рекомендаций. 

Присвоим каждему кластеру свою категорию товаров.

```python
cluster_cat = pd.DataFrame()
categories = ['Текстиль и декор',
              'Комнатные растения',
              'Сумки и тележки',
              'Хозтовары', 
              'Уход и хранение',
              'Товары для кухни',
              'Сад и огород',
              'Товары для ванной',
              'Инструмент и метизы']
for i, category in zip(range(0, 9), categories):
    cluster_cat.loc[i, 'category'] = category
cluster_cat = cluster_cat.reset_index()
cluster_cat.columns = ['cluster_km', 'category']
cluster_cat
```

Создадим характеризующие кластеры метрики: среднее количество заказов, среднее количество приобретенных одним пользователем товаров,  среднее количество позиций товара в заказе, среднюю стоимость товара, средний чек, суммарную выручку. Для этого сгруппируем набор **_final_df_** по кластерам, агрегацию данных пользователей и данных заказов выполним методом **_nunique()_**, агрегацию  данных товара выполним методом **_count()_**, агрегацию данных количества позиций товара и данных общих расходов выполним методом **_sum()_**, агрегацию данных стоимости товара выполним методом **_mean()_**. Результат поместим в переменную **_orders_final_df_**.

```python
orders_final_df = final_df.groupby('cluster_km').agg(
    {'customer_id':'nunique', 
     'order_id':'nunique', 
     'product':'count', 
     'quantity':'sum', 
     'price':'mean',
     'total_value':'sum'}
).reset_index().rename(
    columns={'customer_id':'buyers', 
             'order_id':'orders', 
             'product':'products_in',
             'quantity':'quantitys_in',
             'price':'avg_price',
             'total_value':'revenue'}
)
orders_final_df
```

Добавим в сгруппированный набор столбцы **_'avg_bill'_**, **_'avg_quantitys_per'_** и **_'avg_products_per'_**, в которые поместим значения среднего чека, среднего количества позиций товара в заказе пользователя и среднее значение товара, приходящегося на пользователя.

## Формирование и проверка статистических гипотез

Основной целью любого бизнеса является получение прибыли, наш магазин не является исключением. Так как в наших данных присутствет ключевой показатель -  общая стоимость заказа,  сформулируем гипотезу по проверке статистически значимой разницы в значениях среднего чека в сформированных кластерах покупателей.

Формируем двухстороннюю гипоезу:

* **H₀** - Между исследуемыми выборками  **нет статистически значимой разницы**, значения средних чеков в сформированных кластерах покупателей  **равны** между собой. 
* **H₁** - Между исследуемыми выборками  **есть статистически значимая разница**, значения средних чеков в сформированных кластерах покупателей **имеют различия**.

Для выбора критерия проверки проведем предварительную проверку нормальности распределения значений выборок. каждому из сформированных кластеров покупателей. Предварительно отсортировав набор данных **_final_df_** по возрастанию номеров кластеров.

```python
for cluster in final_df.sort_values(by='cluster_km', ascending=True).cluster_km.unique():
    plt.figure(figsize=(14, 3))
    plt.title('Распределение общей стоимости заказа по кластеру: {}'.format(cluster), fontsize = 16)
    final_df[final_df['cluster_km']==cluster]['total_value'].hist(bins=40, color='grey', alpha=0.7, ec='brown', linewidth=3)
    plt.axvline(np.median(final_df[final_df['cluster_km']==cluster]['total_value']), 
                color='black', linestyle='--', linewidth=1.5, label='median')
    leg = plt.legend(loc='best', fontsize=14)
    for l in leg.legendHandles:
        l.set_linewidth(3)
    plt.grid(True, linestyle=':')
    plt.show()
```

Наблюдаем во всех кластерах пользователей отсутствие нормальности распределения значений общей стоимости заказа, поэтому для проверки гипотез применим непараметрический U-критерий (тест Манна -Уитни).

Определим критический уровень статистической значимости равный 5 %

Создадим список с парами групп.

```python
test_list = list(itertools.combinations([0, 1, 2, 3, 4, 5, 6, 7, 8], 2))
test_list
```

```python
p_data = []
for pair in test_list:
    group1 = pair[0]
    group2 = pair[1]
    pvals = (stats.mannwhitneyu(final_df[final_df['cluster_km']==group1]['total_value'], 
                                final_df[final_df['cluster_km']==group2]['total_value'], alternative='two-sided')[1])
    p_data.append(pvals)
    multitest_result = multipletests(p_data, 0.05, method='sidak')
```

```python
for pair, test_status, corrected_p in zip(test_list, multitest_result[0], multitest_result[1]):
    print('Пара кластеров', pair)
    if test_status:
        print('\033[2;30;47m  Отвергаем нулевую гипотезу \033[0;0m', 'p-значение: ', corrected_p)
    else:
        print('Не получилось отвергнуть нулевую гипотезу', 'p-значение: ', corrected_p) 
    print()
```

Проведенный тест Манна-Уитни показал, что между **30** из 36 пар кластеров нулевые гипотезы **были отвергнуты**, что подтверждает наличие статистически значимых различий значений среднего чека в сформированных кластерах покупателей. Исключение составили шесть пар кластеров:
